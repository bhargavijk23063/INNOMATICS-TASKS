```python
#1.Employee Performance Bonus Eligibility
# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Find the highest performance score
highest_score = max(employees.values())
# Identify all employees who achieved the highest score (handles ties)
top_performers = [name for name, score in employees.items() if score == highest_score]
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
#2. Search Query Keyword Analysis
# Input search query
query = "Buy mobile phone buy phone online"
# Converting the query to lowercase
query = query.lower()
# Remove common punctuation
import string
for p in string.punctuation:
    query = query.replace(p, "")
# Split the query into words
words = query.split()
# Count frequency of each word
from collections import Counter
word_counts = Counter(words)
# Filters only keywords that appear more than once
repeated_keywords = {word: count for word, count in word_counts.items() if count > 1}
print(repeated_keywords)
```

```python
#3.Sensor Data Validation
# List of sensor readings where index = hour, value = reading
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Collect only the even readings along with their hour index
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]
# Show the valid sensor readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#4.Email Domain Usage Analysis
# List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
# Extract domains from email IDs
domains = [email.split("@")[1] for email in emails]
# Count frequency of each domain
from collections import Counter
domain_counts = Counter(domains)
# Calculate percentage usage for each domain
total = len(emails)
domain_percentages = {domain: (count / total) * 100 for domain, count in domain_counts.items()}
# Display results in the required format
for domain, percentage in domain_percentages.items():
    print(f"{domain}: {int(percentage)}%")
```

```python
#5.Sales Spike Detection
# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Calculate average daily sales
average_sales = sum(sales) / len(sales)
# Threshold: 30% above average
threshold = average_sales * 1.3
# Detect days where sales are strictly more than 30% above average
spike_days = [(day+1, value) for day, value in enumerate(sales) if value > threshold]
# Display results
for day, value in spike_days:
    print(f"Day {day}: {value}")
```

```python
#6.Duplicate User ID Detection
# List of user IDs collected during registration
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Count how many times each user ID appears
from collections import Counter
id_counts = Counter(user_ids)
# Filter only those IDs that appear more than once (duplicates)
duplicates = {user_id: count for user_id, count in id_counts.items() if count > 1}
# Print the duplicate IDs along with their counts
for user_id, count in duplicates.items():
    print(f"{user_id} → {count} times")
```
