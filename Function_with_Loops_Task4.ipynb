```python
#1. Mobile Recharge Validation System
# Function to validate recharge amount
def validate_recharge(amount):
    # Check if amount is valid
    return amount >= 50 and amount in [199, 299, 399, 599]
# Loop to allow retry until valid
while True:
    # Take user input
    amount = int(input("Enter recharge amount: ₹"))
    # Validate the amount
    if validate_recharge(amount):
        print("Recharge Successful!")
        break
    print("Invalid amount. Try again.\n")
```

```python
#2. Inventory Reorder Alert System
# Function to check stock and print alert
def check_stock(products):
    # Loop through each product and its quantity
    for product, qty in products.items():
        if qty < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")
# Sample dictionary of products and quantities
inventory = {
    "Apples": 20,
    "Bananas": 10,
    "Oranges": 25,
    "Mangoes": 12
}
# Call function to check inventory
check_stock(inventory)
```

```python
# 3.Student Result Processing System
# Function to calculate average and return result
def process_result(marks):
    # Calculate total marks
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    # Return Pass if average ≥ 50 else Fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
# Sample list of student marks
student_marks = [60, 45, 70, 55, 40]
# Call function
result = process_result(student_marks)
print("Result:", result)
```

```python
# 4.Cab Fare Estimator with Retry Option
# Function to calculate fare
def calculate_fare(km, peak):
    fare = 50 + 12 * km
    if peak:
        fare *= 1.25
    return fare
# Retry until valid input
while True:
    try:
        # Take inputs
        km = float(input("Enter distance in km: "))
        peak = input("Peak hour? (yes/no): ").lower() == "yes"
        # Calculate and print fare
        print(f"Total Fare: ₹{calculate_fare(km, peak):.2f}")
        break
    except:
        print("Invalid input. Try again.\n")
```

```python
#5. Employee Attendance Eligibility Checker
# Function to check eligibility
def check_attendance(attendance):
    # Count present days
    present = attendance.count("P")
    # Calculate percentage
    percentage = (present / len(attendance)) * 100
    # Return eligibility
    return "Eligible" if percentage >= 75 else "Not Eligible"
# Sample attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]
# Call function and print result
print("Attendance Result:", check_attendance(attendance_list))
```

```python
# 6.Password Strength Checker
# Function to check password strength
def check_password(password):
    # Check minimum length
    if len(password) < 8:
        return "Weak Password"
    # Check for digit
    has_digit = any(char.isdigit() for char in password)
    # Check for special character
    has_special = any(char in "@#$" for char in password)
    # Return result
    if has_digit and has_special:
        return "Strong Password"
    return "Weak Password"
# Take input and print result
password = input("Enter password: ")
print(check_password(password))
```
